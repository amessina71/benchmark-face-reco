```python
from clustering import cluster, plot, chinese_whispers, DBSCAN, tsne
import numpy as np
```

# Cluster the faces embedding of a video
1. Load the numpy array containing the embeddings
2. Clusterize
3. Plot TSNE to visualize results

Load the numpy array containg the embeddings

```python
face_embeddings = np.random.rand(43, 128) # 43 embeddings with a 128 shape
```

Clusterize the embeddings with DBSCAN

```python
clusters = cluster(face_embeddings, method=DBSCAN)
```

Plot TSNE and labelled cluster

```python
plot(tsne(face_embeddings), clusters)
```

Clusterize the embeddings with chinese whispers

```python
clusters = cluster(face_embeddings, method=chinese_whispers)
```

Plot

```python
plot(tsne(face_embeddings), clusters)
```
